```python
"""
# Function: 
getEvents()

# Parameters:
csvFile: Path to CSV file from OpenRocket simulation.

# Output:
stateDict: Dictionary with key value pair of flightEvent:IndexInCSV

# Description:
Function outputs a dictionary containing all the flight events that occur in the simulation.
Each flight event has a corresponding index value that indicates which row that event starts
in the output CSV file.
"""
def getEvents(csvFile):

    # Dictionary containing states
    stateDict = {}

    # Read the CSV file and set the first column as the index
    df = pd.read_csv(csvFile, skiprows=3)

    # Iterate through each value in the index (first column) and retrieve flight events
    for index, row in df.iterrows():
        if row[0].startswith('# Event'):
            stateDict[row[0]] = index

    return stateDict
```

```python
# Call getEvents() to our example curve csv file
getEvents('Example Curve.csv')
```

```python
"""
# Function: 
getEventValues()

# Parameters:

# Output:

# Description:

"""
def getEventValues(df, startIndex, endIndex):
        df = pd.read_csv(df, skiprows=3)
        for index, row in df.iterrows():
                if int(index) >= startIndex and int(index) < endIndex:
                        print("index:             ", index)
                        print("Time (s):          ", row[0])
                        print("Altitude (m):      ", row[1])
                        print("Pressure (mbar):   ", row[2])
                else:
                        continue
```

```python
# Call getEventValues() for the event "LAUNCH"
getEventValues("Example Curve.csv", getEvents('Example Curve.csv')["# Event LAUNCH occurred at t=0 seconds"], getEvents('Example Curve.csv')["# Event LIFTOFF occurred at t=0.11 seconds"])
```
